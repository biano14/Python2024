# _ ***Passo 2***: abrir o **database**

Como *Pandas* instalado no ambiente virtual do nosso projeto, agora vamos abrir o ***dataset***. Para isso vamos criar uma *nova* célua de **markdown**, e digitarmos o código-fonte para abrir os dadoas do dataset do notebook. O código-fonte está logo abaixo:


```python
import pandas as pd
    
tabela = pd.read_csv('cancelamentos.csv')
display(tabela)
```

O comando `display()` permite a exibição dos dados como se fosse uma tabela.

Só que agora que finalmente temos os dados armazenados na variável `tabela`, precisaremos filtrar esses dados. Vamos refazer a consulta, mas desta vez iremos retirar a coluna ID, pois ela é inútil para a nossa análise de dados.

A seguir, vamos acrescentar outra célula de código para realizar a filtragem de dados:

```python
# exibe novamente os dados eliminando a coluna id
tabela = tabela.drop('CustomerID', axis=1)
display(tabela)
```
